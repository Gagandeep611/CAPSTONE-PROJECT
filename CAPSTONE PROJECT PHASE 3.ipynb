```python
import sqlite3
```

```python
db=sqlite3.connect('crime_in_india_database.db')
cursor=db.cursor()
```

```python
cursor.execute("CREATE TABLE CRIME_AGAINST_ST(STATE_UT TEXT PRImaryKEY,DISTRICT TEXT,Year INT,MURDER INT,Rape INT,Kidnapping_Abduction INT,Dacoity INT,Robbery INT,Arson INT,Hurt INT,Protection_of_Civil_Rights_Act INT,Prevention_of_atrocitie_Act INT,Other_Crimes_Against_STs INT)")
```

```python
with open('02_District_wise_crimes_committed_against_ST_2001_2012.csv','r')as file:
    no_records=0
    for row in file:
        cursor.execute("INSERT INTO CRIME_AGAINST_ST VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?)",row.split(","))
        db.commit()
        no_records +=1
        
print(no_records, 'RECORDS INSERTED')
```

```python
results = cursor.execute("select * from CRIME_AGAINST_ST")
for i in results:
    print(i)
```

```python
result = cursor.execute("SELECT MAX(Dacoity) FROM CRIME_AGAINST_ST")
print('MAXIMUM DACOITY =',result.fetchall())
```

```python
result = cursor.execute("SELECT MAX(Robbery) FROM CRIME_AGAINST_ST")
print('MAXIMUM Robbery =',result.fetchall())
```

```python
result=cursor.execute("SELECT DISTRICT, YEAR, MURDER FROM CRIME_AGAINST_ST ORDER BY MURDER")
for i in result:
    print(i)
```

```python
result = cursor.execute("SELECT MIN(Murder) FROM CRIME_AGAINST_ST")
print('MAXIMUM Robbery =',result.fetchall())
```
